```python
import matplotlib.pyplot as plt
import numpy as np
import random
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from colorama import Fore, Style, init
```

```python
# Initialize colorama
init(autoreset=True)
```

```python
class CodeGeneratorAgent:
    def __init__(self, model, tokenizer, max_retries=3):
        self.model = model
        self.tokenizer = tokenizer
        self.max_retries = max_retries

    def generate_code(self, prompt, max_length=300):
        inputs = self.tokenizer(prompt, return_tensors="pt")
        outputs = self.model.generate(**inputs, max_length=max_length, pad_token_id=self.tokenizer.eos_token_id)
        generated_code = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Filter out non-code text from the generated output
        code_lines = generated_code.split('\n')
        filtered_code = []
        for line in code_lines:
            line = line.strip()  # Remove leading and trailing whitespace
            if not line or line.startswith("#") or line.startswith(">>>") or '"""' in line or "print(" in line:
                continue
            filtered_code.append(line)

        return '\n'.join(filtered_code)

    def execute_code(self, code):
        exec_locals = {}
        try:
            exec(code, globals(), exec_locals)
        except SyntaxError as se:
            exec_locals['error'] = f"SyntaxError: {se}"
        except Exception as e:
            exec_locals['error'] = str(e)
        return exec_locals

    def run(self, prompt):
        for attempt in range(self.max_retries):
            generated_code = self.generate_code(prompt)
            print(Fore.GREEN + f"Attempt {attempt + 1} Generated Code:\n{generated_code}")
            exec_locals = self.execute_code(generated_code)
            if 'error' not in exec_locals:
                return exec_locals
            print(Fore.RED + f"Error during execution: {exec_locals['error']}")
        return {'error': 'Failed to generate and execute code successfully after multiple attempts.'}

if __name__ == "__main__":
    prompt = """
    Write a Python script to:
    1. Generate two random numbers.
    2. Add them together.
    3. Print the result.
    """

    agent = CodeGeneratorAgent(model, tokenizer)
    exec_locals = agent.run(prompt)

    # Example usage of the generated code
    if 'error' in exec_locals:
        print(Fore.RED + f"Final error: {exec_locals['error']}")
```
